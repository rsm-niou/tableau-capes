```python
salary_dat['Gross Pay'] = salary_dat['Gross Pay'].str.replace(',', '').astype(float)
salary_dat['Regular Pay'] = salary_dat['Regular Pay'].str.replace(',', '').astype(float)
salary_dat['Overtime Pay'] = salary_dat['Overtime Pay'].str.replace(',', '').astype(float)
salary_dat['Other Pay'] = salary_dat['Other Pay'].str.replace(',', '').astype(float)
```
